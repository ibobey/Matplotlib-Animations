```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("population_pyramid_1950-2022.csv")
```

```python
data_turkey = data[data.Country == 'Turkey']
```

```python
data_turkey.reset_index(drop=True,inplace=True)
```

```python
del data
```

```python
x_ =list(data_turkey.Age.unique())
x = [*range(1,len(x_)+1)]
y = data_turkey[data_turkey.Year == 2022]["M"]
y2 = -data_turkey[data_turkey.Year == 2022]["F"]
```

```python
sns.set_theme(
    context='paper',
    style='ticks',
    palette='pastel')
```

```python
fig,ax = plt.subplots(figsize=(12.5,10))

M = ax.barh(x,y,alpha=0.75,color="#87CEFA")
F = ax.barh(x,y2,alpha=0.75,color="#FFC0CB")


ax.scatter(y,x,c="#FF0000")
ax.scatter(y2,x,c="#FF0000")

ax.bar_label(M,labels=[str(abs(i))  for i in y],label_type="edge",padding=8)
ax.bar_label(F,labels=[str(abs(i))  for i in y2],label_type="edge",padding=8)


ax.axvline(0,color="#FF0000",ls="-",alpha=0.35)

ax.set_yticks(x,x_)


ax.set_xticks([])

ax.set_xlim(-4e6,4e6)


sns.despine(left=True)

ax.grid(True,alpha=0.45,ls="-.")

plt.title(f"Population of Turkey {6}",fontdict={"fontsize":18})
plt.tight_layout()
plt.show()
```
